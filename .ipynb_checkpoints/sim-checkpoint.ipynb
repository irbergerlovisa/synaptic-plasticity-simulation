```python
# In order of & with notation of paper Tully et al 2014.

# Neuron model, parameters: -------------------------

C_m = 250 * pF #membrane capacity
g_L = 16.67 * nS #leak conductance
E_L = -70 * mV #leak reversal potential

E_inh = -75 * mV #inhibitory reversal potential, I will not use exhibitory? 
E_ex = 0 * mV #excitatory reversal potential

unknown = 50 * pA #current scaling factor,-- what is this letter?
V_th = -55 * mV #membrane voltage threshold
V_res = -60 * mV #membrane reset potential
t_ref = 2 * ms #refractory period 
s_dt = 0.1 * ms #time resolution

# Channel model, parameters: -------------------------

tau_z = 10 * ms # Z trace time constant, I put tau_z since tau_zj = tau_zi (?)
tau_e = 100 * ms # E trace time constant
tau_p = 10000 * ms # P trace time constant
# S_j spike recordings directly in Brian, & otherwise..?
f_max = 20 * Hz # highest firing rate
epsilon = 1/(f_max * tau_p) * Hz #lowest firing rate
t_spike = 0.1 * ms # spike duration

# Synapse model, parameters: -------------------------

# we already have tau_zi (=tau_z), tau_e, tau_p
g_max = 2.0 * nS # synaptic peak conductance
tau_ex = 0.2 * ms # synaptic rise time, for excitatory input, (alpha rise time(?), synapse model "conductance based w. alpha-shaped PSC").
tau_inh = 2.0 * ms # synaptic rise time, for inhibitory input.
d = 0.1 * ms # transmission delay = t_spike = dt, alter to this.
kappa = 1.0 # 0 to freeze plasticity, so on/off
```

```python
# Neuron model, defining equations:

eqs_neurons = '''

# Memory traces for postsynaptic neuron j: will add same for presynaptic i later.

dS_j/dt = -S_j/s_dt : 1
dZ_j/dt = (S_j / (f_max * t_spike) - Z_j + epsilon * second ) / tau_z : 1
dE_j/dt = (Z_j - E_j) / tau_e : 1
dP_j/dt = (kappa * (E_j - P_j)) / tau_p : 1

# Bias 
beta = log(clip(P_j, min_num, inf)) : 1 # beta = log(P_j), but if P_j < min_num, clip sets P_j <- min_num.
I_beta = phi * beta : amp # written as phi_{I beta} in the paper, but should actually be I_beta.

# On-switch: 
# use this? try without?

# Total voltage:
g_ex : siemens # sum for all excitatory synapses
g_inh : siemens # sum for all inhibitory synapses

# LIF Neuron Model:

dV/dt = (g_L * (V - E_L) + g_ex * (V - E_ex) + g_inh * (V - E_inh) + I_beta) / (- C_m) : volt (unless refractory)

'''

neurons = NeuronGroup(2, model = eqs_neurons, method = 'euler', threshold = 'V >= V_th', reset = 'V = V_res', refractory = t_ref)
neurons.V = -70 * mV # leak reversal potential
# have not added the b_on part

# I am on line 44

# synapse model, look at Brian documentation
```
